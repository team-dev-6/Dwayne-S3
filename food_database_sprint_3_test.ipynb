```python
# Use the "INSERT INTO" statement to fill a table in MySQL

# Addding multiple rows to my table with new values using executemany

mycursor = conn.cursor()

sql = "INSERT INTO fooddb.food_table_woolworths1 (Item_No, Food_Item, Category, Size, Price, Stock_Quantity, Stock_Value, Expiry_Date, Reorder_Time_In_Days, Quantity_In_Reorder) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
val = [
    ("A041", "Double Ply Super Absorbent Roller Towel", "Household & Cleaning", "2Pk", "22.99", "14", "321.86", "2024/05/19", "3", "18"),
    ("A042", "Bulk Super Soft 2 Ply Toilet Paper", "Household & Cleaning", "18 Pk", "109.99", "8", "879.92", "2025/11/06", "5", "24")
]
mycursor.executemany(sql, val)

conn.commit()

print(mycursor.rowcount, "record inserted.")
```

```python
# You can delete records from an existing table by using the "DELETE FROM" statement

# Deleting mutiple rows using OR

mycursor = conn.cursor()

sql = "DELETE FROM fooddb.food_table_woolworths1 WHERE Item_No = 'A041' OR Item_No = 'A042'"

mycursor.execute(sql)

conn.commit()

print(mycursor.rowcount, "record(s) deleted")
```
